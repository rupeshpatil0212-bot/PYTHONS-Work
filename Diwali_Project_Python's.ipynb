**Number's of Total Revenue by the Gender**............

```python
# Total Amount Spent by Gender
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_gen, x='Gender', y='Amount', palette='viridis')
plt.title('Total Revenue by Gender')
plt.show()
```

**1. Why did you pick this specific chart?**

 The Bar Plot was chosen because it is the most effective tool for comparing a numerical variable (Amount) across different categorical groups (Gender).

 **2. What is/are the insight(s) found in the chart?**

Primary Spenders: It identifies which gender segment contributes more to the total revenue or has a higher average transaction value.

Gap Analysis: It reveals the magnitude of the spending difference, showing whether one group significantly outperforms the other or if spending is relatively balanced.

**3. Will the insights help a positive business impact?**

Targeted Marketing: Allocating advertising budget toward the higher-spending demographic to maximize ROI.

Product Strategy: Tailoring product inventory or features to better suit the preferences of the dominant consumer group.

```python
# Correlation Heatmap visualization code
plt.figure(figsize=(15,10))
sns.heatmap(df.select_dtypes(include=np.number).corr(),annot=True)
plt.title('Correlation of the columns')
plt.show()
```

**1. Why was this specific chart picked?**

A Heatmap (or Correlation Matrix) is the standard choice because it visualizes the relationship between multiple variables simultaneously. It uses a color-coded scale to represent the strength of a relationship, making it easy to spot patterns across a large dataset that would be impossible to see in a raw table.

**2. What are the insights found in the chart?**

Direction: You can see if variables have a positive relationship (move together) or a negative relationship (one goes up, the other goes down).

Strength: The intensity of the color reveals how closely linked the variables are (e.g., $1.0$ is a perfect match).

Redundancy: It identifies "Multicollinearity," where two columns provide the same information, allowing you to simplify your data.

**3. Will the gained insights help a positive business impact?**

Resource Optimization: Identifying which specific factors (like "Marketing Spend") actually drive "Revenue," allowing you to stop wasting money on low-impact activities.

Model Accuracy: Improving the performance of predictive tools by removing "noise" and redundant data.

Risk Mitigation: Spotting hidden dependencies that could cause a business process to fail if one variable changes unexpectedly.
